```python
X = pd.get_dummies(df)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_x = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=2)
```

```python
cluster_lab = model.fit_predict(scaled_x)
```

```python
X['Cluster'] = cluster_lab
```

```python
X.head()
```

```python
plt.figure(figsize=(12,8), dpi=140)
X.corr()['Cluster'].sort_values().plot(kind='bar')
```

```python
ssd = []

for k in range(2,10):

    model = KMeans(n_clusters=k)
    model.fit(scaled_x)

    ssd.append(model.inertia_)
```

```python
ssd
```

```python
plt.plot(range(2,10), ssd, 'o--')
```

```python
pd.Series(ssd)
```

```python
pd.Series(ssd).diff()
```

```python
import matplotlib.image as mimg
import matplotlib.pyplot as plt
```

```python
image_array = mimg.imread('../data/palm_trees.jpg')
```

```python
image_array
```

```python
plt.imshow(image_array)
```

```python
(h,w,c) = image_array.shape
```

```python
c
```

```python
imagr_array_2d = image_array.reshape(h*w,c)
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=6)
```

```python
labels = model.fit_predict(imagr_array_2d)
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int)
```

```python
rgb_codes
```

```python
rgb_codes[labels]
```

```python
import numpy as np
```

```python
quant_image = np.reshape(rgb_codes[labels], (h,w,c))
```

```python
plt.figure(figsize=(12,10),dpi=150)
plt.imshow(quant_image)
```
