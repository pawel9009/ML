```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/cluster_mpg.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df['origin'].value_counts()
```

```python
df_dummies = pd.get_dummies(df.drop('name',axis=1))
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sclaer = MinMaxScaler()
```

```python
sclaed_data = sclaer.fit_transform(df_dummies)
```

```python
scaled_df = pd.DataFrame(sclaed_data, columns=df_dummies.columns)
```

```python
scaled_df
```

```python
plt.figure(figsize=(12,10),dpi=100)
sns.heatmap(scaled_df, cmap='viridis')
```

```python
plt.figure(figsize=(12,10),dpi=100)
sns.clustermap(scaled_df)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=3)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
plt.figure(figsize=(10,5), dpi=100)
sns.scatterplot(data=df, x='mpg', y='weight' , hue= cluster_labels, palette='viridis')
```

```python
model = AgglomerativeClustering(n_clusters=None, distance_threshold=0)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
```

```python
linkage_matrix = hierarchy.linkage(model.children_)
```

```python
linkage_matrix
```

```python
plt.figure(figsize=(20,12))
dendro = dendrogram(linkage_matrix, truncate_mode='lastp', p=10)
```
