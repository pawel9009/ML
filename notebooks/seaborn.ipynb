```python
df = pd.read_csv('../data/StudentsPerformance.csv')

df.head()
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.boxplot(data=df, y='math score', x='parental level of education',hue='test preparation course')

plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
plt.figure(figsize=(15,15), dpi=200)
sns.violinplot(data=df, x='math score', y='parental level of education',hue='test preparation course', split=True)
```

```python
plt.figure(figsize=(10,5), dpi=200)
sns.swarmplot(data=df, x='math score',y='gender' ,size=2, hue='test preparation course', dodge=True)
plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
sns.boxenplot(data=df, x='math score', y='test preparation course', hue='gender')
```

```python
df.head()
```

```python
sns.jointplot(data=df, x='math score', y='reading score', kind = 'hist', alpha=0.2, hue='gender')
```

```python
sns.pairplot(data=df, hue='gender')
```

```python
df.head()
```

```python
sns.catplot(data=df, x='gender', y='math score', kind = 'box',
            col = 'lunch')
```

```python
g=sns.PairGrid(df, hue='gender')

g= g.map_upper(sns.scatterplot)

g= g.map_diag(sns.histplot)

g = g.map_lower(sns.kdeplot)

g = g.add_legend()
```

```python
df = pd.read_csv('../data/application_record.csv')
```

```python
df.head()
```

```python
employed = df[df['DAYS_EMPLOYED']<0]
```

```python
employed['DAYS_EMPLOYED'] = employed['DAYS_EMPLOYED']*-1
```

```python
employed['DAYS_BIRTH'] = employed['DAYS_BIRTH']*-1
```

```python
plt.figure(figsize=(10,5), dpi=200)
sns.scatterplot(data=employed, y='DAYS_EMPLOYED', x='DAYS_BIRTH', alpha=0.01, linewidth=0)
```

```python
df['YEARS'] = -1*df['DAYS_BIRTH']/365
sns.histplot(data=df, x='YEARS', bins=48, color='red')
```
