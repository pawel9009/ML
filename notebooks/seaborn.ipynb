```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot  as plt
```

```python
df = pd.read_csv('../data/dm_office_sales.csv')
```

```python
df
```

```python
sns.scatterplot(data=df, x='salary', y='sales', s=200,alpha=0.7, style='level of education', hue='level of education')
```

```python
plt.figure(figsize=(10,5))
sns.rugplot(x='salary', data=df, height=0.5)
```

```python
sns.set(style='dark')
sns.displot(data=df, x='salary', bins=30, color='red', edgecolor='blue')
```

```python
sns.histplot(data=df, x='salary', kde=True)
```
