```python
m = np.linspace(0,10,11)
```

```python
c = 3*10**8
```

```python
c
```

```python
E = m* c**2
```

```python
E
```

```python
plt.plot(m,E,color='red',lw=5)
plt.title('E=mc^2')
plt.xlabel('M in g')
plt.ylabel('E in J')
plt.xlim(0,10)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(m,E,color='red', lw=5)
```

```python
labels = ['1 mo', '3 mo', '6 mo', '1 yr', '2 yr', '3 yr', '5 yr', '7yr', '10yr', '20yr', '25 yr']

july16_2007 = [4.75,4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]

july16_2020 = [0.12,0.11, 0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(labels,july16_2007, label ='2007')
ax.plot(labels,july16_2020, label ='2020')
ax.legend()
```

```python
fig,ax = plt.subplots(nrows=2,ncols=1, figsize=(12,8))

ax[0].plot(labels,july16_2007, label ='2007')
ax[0].set_title('2007')

ax[1].plot(labels,july16_2020, label ='2020')
ax[1].set_title('2020')
```
